```python
import pandas as pd 

df = pd.read_csv('Scrape.csv')
df.shape
```

```python
# Limpar a coluna Area (remover texto e converter para float)
df['Area_num'] = df['Area'].str.replace(' m²', '').str.replace('.', '').str.replace(',', '.').astype(float)

# Calcular Preco medio (preço por m²)
df['Preco_medio'] = df['Preco'] / df['Area_num']

df['Preco_medio'].head()
```

```python
# Verificar os cálculos - mostrando Preco, Area, Area_num e Preco_medio
df[['Nome', 'Preco', 'Area', 'Area_num', 'Preco_medio']].head(10)
```

```python
# Formatar Preco_medio para melhor visualização
df['Preco_medio'] = df['Preco_medio'].round(2)

# Ver resultado formatado
df[['Nome', 'Endereco', 'Preco', 'Area', 'Preco_medio']].head()
```
